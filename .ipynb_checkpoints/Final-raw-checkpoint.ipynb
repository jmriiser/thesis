```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import umap
import time
from sklearn.pipeline import Pipeline

from utils import plot_missing_percentages, drop_missing_cols
from pipelines import preprocessor, umap_preprocessor, pca_preprocessor, get_data, custom_eval, dimension_reducer, DIAGCOLORS, plot_clusters, plot_b2b
```

### Load Data

get_data loads the data set, and performs some data cleaning (i.e. rename, replace missing values with 'unknown', etc.)

```python
data = get_data()
data = drop_missing_cols(data, 60)
plot_missing_percentages(data)
```

#### Holdout Data set

Create a holdout data set by modifying the original data. Removes some of the key features and reduces its values.

```python
holdout_al = data[data['PHDIAG'] == 'AL amyloidose']
holdout_al = holdout_al.assign(DPD='Ukjent')
holdout_al['HEBNPRE']-=200

X_al = holdout_al.drop(columns=['PHDIAG', 'DMSEX'])
y_al = holdout_al['PHDIAG']
```

```python
holdout_attr = data[data['PHDIAG'] == 'TTR amyloidose']
holdout_attr = holdout_attr.assign(DPD='Ukjent')
holdout_attr = holdout_attr.assign(PHMH2=0)
holdout_attr = holdout_attr.assign(PHRVSPYN='Nei')

holdout_attr['HEBNPRE']-=100

X_attr = holdout_attr.drop(columns=['PHDIAG', 'DMSEX'])
y_attr = holdout_attr['PHDIAG']
```

```python
holdout_other = data[data['PHDIAG'] == 'Annet']
holdout_other = holdout_attr.assign(DPD='Ukjent')
holdout_other = holdout_attr.assign(PHMH2=0)
holdout_other = holdout_attr.assign(PHRVSPYN='Nei')

holdout_other['HEBNPRE']-=100

X_other = holdout_other.drop(columns=['PHDIAG', 'DMSEX'])
y_other = holdout_other['PHDIAG']
```

### Data Preparation

Thee preprocessor applies the preprocessing methods on the data set. 
- X_preprocessed returns the data projected in 2D by t-SNE.
- X_umap returns the data projected in 2D by UMAP.

We decided to remove DMSEX (i.e. gender) to the data as it causes issues due to covariance and strong bias.

```python
X = data
y = data['PHDIAG']
```

```python
X_preprocessed = preprocessor.fit_transform(X.drop(columns=['PHDIAG', 'DMSEX']))
X_umap = umap_preprocessor.fit_transform(X.drop(columns=['PHDIAG', 'DMSEX']))
```

```python
X_al_proc = pca_preprocessor.transform(X_al)
X_attr_proc = pca_preprocessor.transform(X_attr)
X_other_proc = pca_preprocessor.transform(X_other)
```

### Parameter Optimization

#### Selecting Epsilon for DBSCAN

```python
from sklearn.neighbors import NearestNeighbors

neigh = NearestNeighbors(n_neighbors=2)
nbrs = neigh.fit(X_preprocessed)

distances, indices = nbrs.kneighbors(X_preprocessed)
distances = np.sort(distances, axis=0)
distances = distances[:,1]

plt.plot(distances)
```

```python
neigh = NearestNeighbors(n_neighbors=2)
nbrs = neigh.fit(X_umap)

distances, indices = nbrs.kneighbors(X_umap)
distances = np.sort(distances, axis=0)
distances = distances[:,1]

plt.plot(distances)
```

### The Clustering Models

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, SpectralClustering
import hdbscan

num_clusters = 4
num_clusters_umap = 4

models = [
    KMeans(n_clusters=num_clusters, init="k-means++", n_init=50, max_iter=500, random_state=0),
    AgglomerativeClustering(n_clusters=num_clusters, affinity='euclidean', linkage='ward'),
    hdbscan.HDBSCAN(min_cluster_size=7),
    DBSCAN(eps=.85, min_samples=3),
    DBSCAN(eps=.35, min_samples=3),
]
```

## K-Means Clustering

```python
kmeans = models[0]

kmeans.fit(X_preprocessed)
kmeans_pred = kmeans.predict(X_preprocessed)

kmeans_al_pred = kmeans.predict(X_al_proc)
kmeans_attr_pred = kmeans.predict(X_attr_proc)
kmeans_other_pred = kmeans.predict(X_other_proc)

X['kmeans'] = kmeans.labels_

centroids = kmeans.cluster_centers_

kmeans_umap = models[0]

kmeans_umap.fit(X_umap)
kmeans_pred_umap = kmeans.predict(X_umap)
centroids_umap = kmeans_umap.cluster_centers_
X['kmeans_umap'] = kmeans_umap.labels_


plt.figure(figsize=(24, 10))

plt.subplot(1, 2, 1)
sc = plt.scatter(X_preprocessed[:,0], X_preprocessed[:,1], c=kmeans_pred)
plt.scatter(centroids[:,0] , centroids[:,1] , s = 80, marker='x', color='orange') 
plt.scatter(X_al_proc[:,0], X_al_proc[:,1], marker='s', c='red')
plt.scatter(X_attr_proc[:,0], X_attr_proc[:,1], marker='*', c='red')
plt.scatter(X_other_proc[:,0], X_other_proc[:,1], marker='+', c='red')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.legend(*sc.legend_elements(), title='clusters')

plt.subplot(1, 2, 2)
sc_umap = plt.scatter(X_umap[:,0], X_umap[:,1], c=kmeans_pred_umap)
plt.scatter(centroids_umap[:,0] , centroids_umap[:,1] , s = 80, marker='x', color='orange') 
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')
plt.legend(*sc_umap.legend_elements(), title='clusters')

plt.savefig('kmeans')
plt.show()

kmeans_eval = custom_eval(X_preprocessed, y, kmeans_pred)
kmeans_umap_eval = custom_eval(X_umap, y, kmeans_pred_umap)
```

```python
from sklearn.metrics import silhouette_score 
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score

def eval_all_metrics(X_preprocessed, pred, X_umap, pred_umap):
    # Calculate cluster validation metrics
    score_kemans_s = silhouette_score(X_preprocessed, pred, metric='euclidean')
    score_kemans_c = calinski_harabasz_score(X_preprocessed, pred)
    score_kemans_d = davies_bouldin_score(X_preprocessed, pred)

    score_kemans_u_s = silhouette_score(X_umap, pred_umap, metric='euclidean')
    score_kemans_u_c = calinski_harabasz_score(X_umap, pred_umap)
    score_kemans_u_d = davies_bouldin_score(X_umap, pred_umap)

    print('Silhouette Score: %.4f' % score_kemans_s)
    print('Calinski Harabasz Score: %.4f' % score_kemans_c)
    print('Davies Bouldin Score: %.4f' % score_kemans_d)

    print('Silhouette Score: %.4f' % score_kemans_u_s)
    print('Calinski Harabasz Score: %.4f' % score_kemans_u_c)
    print('Davies Bouldin Score: %.4f' % score_kemans_u_d)
    
eval_all_metrics(X_preprocessed, kmeans_pred, X_umap, kmeans_pred_umap)
```
