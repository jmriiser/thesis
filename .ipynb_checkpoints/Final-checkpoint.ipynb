```python
import os
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import umap
import time
from sklearn.pipeline import Pipeline

from utils import plot_missing_percentages, drop_missing_cols
from pipelines import preprocessor, umap_preprocessor, get_data, custom_eval, dimension_reducer, DIAGCOLORS, plot_clusters, plot_b2b
```

```python
data = get_data()
data = drop_missing_cols(data, 60)
plot_missing_percentages(data)
```

```python
X = data.drop(columns=['PHDIAG'])
y = data['PHDIAG']
```

# Models

```python
from sklearn.cluster import KMeans, AgglomerativeClustering, DBSCAN, SpectralClustering
import hdbscan

num_clusters = 3

models = [
    KMeans(n_clusters=num_clusters, init="k-means++", n_init=50, max_iter=500, random_state=0),
    AgglomerativeClustering(n_clusters=num_clusters, affinity='euclidean', linkage='ward'),
    hdbscan.HDBSCAN(min_cluster_size=10),
    DBSCAN(),
    SpectralClustering(n_clusters=num_clusters)
]
```

```python
X_preprocessed = preprocessor.fit_transform(X)
X_umap = umap_preprocessor.fit_transform(X)
```

```python
print(X_preprocessed[0])
print(X_umap[0])
```

### Selecting number of clusters

```python
from matplotlib.lines import Line2D
legend_elements = [Line2D([0], [0], marker='o', color='w', label='AL',
                          markerfacecolor='red', markersize=10),
                   Line2D([0], [0], marker='o', color='w', label='ATTR',
                          markerfacecolor='blue', markersize=10),
                   Line2D([0], [0], marker='o', color='w', label='Other',
                          markerfacecolor='green', markersize=10)]
```

## K-Means Clustering

```python
kmeans = models[0]

kmeans.fit(X_preprocessed)
kmeans_pred = kmeans.predict(X_preprocessed)
#X_preprocessed['kmeans'] = list(map(lambda x: [x], kmeans.fit_predict(X_preprocessed)))
centroids = kmeans.cluster_centers_

kmeans_umap = models[0]

kmeans_umap.fit(X_umap)
kmeans_pred_umap = kmeans.predict(X_umap)
centroids_umap = kmeans_umap.cluster_centers_

#plt.style.use('ggplot')
plt.figure(figsize=(18, 7))

plt.subplot(1, 2, 1)
plt.scatter(X_preprocessed[:,0], X_preprocessed[:,1], c=kmeans_pred)
plt.scatter(centroids[:,0] , centroids[:,1] , s = 80, marker='x', color='orange') 
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')

plt.subplot(1, 2, 2)
plt.scatter(X_umap[:,0], X_umap[:,1], c=kmeans_pred_umap)
plt.scatter(centroids_umap[:,0] , centroids_umap[:,1] , s = 80, marker='x', color='orange') 
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')

plt.savefig('kmeans')
plt.show()

kmeans_eval = custom_eval(X_preprocessed, y, kmeans_pred)
kmeans_umap_eval = custom_eval(X_umap, y, kmeans_pred_umap)
```

```python
plt.figure(figsize=(18, 7))

plt.subplot(1, 2, 1)
plt.scatter(X_preprocessed[:,0], X_preprocessed[:,1], c=y.map(DIAGCOLORS))
plt.scatter(centroids[:,0] , centroids[:,1] , s = 80, marker='x', color='orange')
plt.xlabel('t-SNE 1')
plt.ylabel('t-SNE 2')
plt.legend(handles=legend_elements)

plt.subplot(1, 2, 2)
plt.scatter(X_umap[:,0], X_umap[:,1], c=y.map(DIAGCOLORS))
plt.scatter(centroids_umap[:,0] , centroids_umap[:,1] , s = 80, marker='x', color='orange')
plt.xlabel('UMAP 1')
plt.ylabel('UMAP 2')

plt.legend(handles=legend_elements)
plt.show()
plt.savefig('ground_truth')
```

```python
#plot_clusters(X_preprocessed, KMeans, (), {'n_clusters':5, 'init':"k-means++", 'n_init':50, 'max_iter':500, 'random_state':0})
```

```python
from sklearn.metrics import silhouette_score 
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score
# Fit K-Means
kmeans_1 = models[0]
# Use fit_predict to cluster the dataset
predictions = kmeans_1.fit_predict(X_preprocessed)
# Calculate cluster validation metrics
score_kemans_s = silhouette_score(X_preprocessed, kmeans_1.labels_, metric='euclidean')
score_kemans_c = calinski_harabasz_score(X_preprocessed, kmeans_1.labels_)
score_kemans_d = davies_bouldin_score(X_preprocessed, predictions)


print('Silhouette Score: %.4f' % score_kemans_s)
print('Calinski Harabasz Score: %.4f' % score_kemans_c)
print('Davies Bouldin Score: %.4f' % score_kemans_d)
```

## Agglomerative Hierarchical Clustering

```python
agglomerative = models[1]
agglo_preds = agglomerative.fit_predict(X_preprocessed)

agglo_umap = models[1]
agglo_preds_umap = agglo_umap.fit_predict(X_umap)

plot_b2b(X_preprocessed, X_umap, agglo_preds, agglo_preds_umap, save='agglo')

agglo_eval = custom_eval(X_preprocessed, y, agglo_preds)
agglo_umap_eval = custom_eval(X_umap, y, agglo_preds_umap)
```

## HDBSCAN

```python
hdbsc = models[2]
hdbscan_preds = hdbsc.fit_predict(X_preprocessed)

hdbsc_umap = models[2]
hdbscan_preds_umap = hdbsc.fit_predict(X_umap)

plot_b2b(X_preprocessed, X_umap, hdbscan_preds, hdbscan_preds_umap, save='hdbscan')

hdbscan_eval = custom_eval(X_preprocessed, y, hdbscan_preds)
hdbscan_umap_eval = custom_eval(X_umap, y, hdbscan_preds_umap)
```
